```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
covid = "data_sources2/Counties.csv"
rallies = "data_sources2/pres_rallies_2020.csv"
```

```python
# Creating data frames
covid_df = pd.read_csv(covid, encoding="ISO-8859-1")
rallies_df = pd.read_csv(rallies, encoding="ISO-8859-1")
```

```python
#Viewing data frames
covid_df.head()
```

```python
#chekcing data
covid_df.count()
```

```python
# Using GroupBy in order to separate the data into fields according to combine county/state values
grouped_covid_df = covid_df.groupby(['county_state'])

# In order to be visualized, a data function must be used...
grouped_covid_df.describe().head(10)
```

```python
#correcting data types 
covid_df["cases"] = covid_df.loc[:, "cases"].astype(float)
covid_df["deaths"] = covid_df.loc[:, "deaths"].astype(float)
covid_df["Date2"] = covid_df["Date2"].astype('datetime64[ns]')
```

```python
covid_df.info()
```

```python
#extracting the year and month name and adding them as a columns, keeping county/state, date, and # of cases
covid_by_month_df =  (covid_df.groupby([covid_df['county_state'], covid_df['Date2'].dt.year.rename('year'), 
                   covid_df['Date2'].dt.month_name().rename('month'),covid_df['date1']]) ['cases'].sum().reset_index())
covid_by_month_df.head()
```

```python
#Ordering new df by date

covid_by_month_df = covid_by_month_df.sort_values(["county_state", "date1"])
covid_by_month_df.head()
```

```python
# using loc to pull just the covid data for election rally months
aug_sep_oct_nov = covid_by_month_df.loc[(covid_by_month_df["month"] == "August") | (
    covid_by_month_df["month"] == "September") | (
    covid_by_month_df["month"] == "October") | (
    covid_by_month_df["month"] == "November"), :]
aug_sep_oct_nov
```

```python
#finding the begining of month and end of month cases for each county/state 

covid_min_cases = pd.DataFrame(aug_sep_oct_nov.groupby(['county_state', 'month']).min())
covid_max_cases = pd.DataFrame(aug_sep_oct_nov.groupby(['county_state', 'month']).max())
```

```python
# adding the end value of cases for each month in to DF w begining data 
end_cases = covid_max_cases['cases']
covid_min_cases['end_cases'] = end_cases

covid_min_cases
```

```python
#county_state = covid_max_cases['county_state']
#covid_min_cases['county_and_state'] = county_state
#covid_min_cases
```

```python
#finding the % change in cases fron begining to end of month
perc_cng_cases = (covid_min_cases["end_cases"] - covid_min_cases["cases"]) / covid_min_cases["cases"]
covid_min_cases["perc_cng_cases"] = perc_cng_cases
covid_min_cases
```

```python
#correcting the oreder of months
covid_cleaned = covid_min_cases.sort_values(["county_state", "date1"])
covid_cleaned
```

```python
# Export file as a CSV, without the Pandas index, but with the header
covid_cleaned.to_csv("data_sources2/covid_ave_growth_per_month.csv", index=False, header=True)
```
