```python
import pandas as pd
```

```python
# Store csv data into DataFrame
csv_file = "./data_sources/combined_data_clean.csv"
df = pd.read_csv(csv_file)

# Convert percent to float
df['average_growth_new_cases'] = df['average_growth_new_cases'].str.rstrip('%').astype('float')
df['post_rally_growth_new_cases'] = df['post_rally_growth_new_cases'].str.rstrip('%').astype('float')

df.head()
```

```python
df['post_rally_change_in_growth'] = df['post_rally_growth_new_cases'] - df['average_growth_new_cases']
df['post_rally_change_in_growth'] = df['post_rally_change_in_growth'] / abs(df['average_growth_new_cases'])


df['average_growth_new_cases'] = df['average_growth_new_cases']/100
df['post_rally_growth_new_cases'] = df['post_rally_growth_new_cases']/100

df['average_growth_new_cases'] = df['average_growth_new_cases'].astype(float).map("{:.2%}".format)
df['post_rally_growth_new_cases'] = df['post_rally_growth_new_cases'].astype(float).map("{:.2%}".format)
df['post_rally_change_in_growth'] = df['post_rally_change_in_growth'].astype(float).map("{:.2%}".format)

df.head()

df.to_csv("data_sources/combined_data_clean.csv", index=False, header=True)
```
